1 imdb_df['crew'].isna()
- Checks which rows in the crew column are NaN (null/missing).
- Returns a boolean Series (True for null, False for non-null).

2 ~imdb_df['crew'].isna()
- The ~ operator flips the boolean values.
- So now it’s True for non-null values and False for null values.
- This ensures we only target rows where crew is not null.

3 imdb_df.loc[condition, 'crew']
- .loc is used for label-based indexing.
- Here, it selects only the rows where crew is non-null, and specifically the crew column.

4 .apply(convert_to_list)
- Applies the function convert_to_list to each non-null value in the crew column.
- Example: if crew contains a string like "John, Mary", your function might turn it into ["John", "Mary"].

5 Assignment (=)
- The transformed values are reassigned back into the same crew column, but only for the non-null rows.
- Null values remain untouched.


```python
#apply the above function only on non null values in genres column
imdb_df.loc[~imdb_df['genres'].isna(),'genres']= imdb_df.loc[~imdb_df['genres'].isna(),'genres'].apply(convert_string_to_list)
```

```python
#apply the above function only on non null values in cast column
imdb_df.loc[~imdb_df['cast'].isna(),'cast']= imdb_df.loc[~imdb_df['cast'].isna(),'cast'].apply(convert_string_to_list)
```

```python
#apply the above function only on non null values in crew column
imdb_df.loc[~imdb_df['crew'].isna(),'crew']= imdb_df.loc[~imdb_df['crew'].isna(),'crew'].apply(convert_string_to_list)
```

#Q4.Make a table of all the producers and directors of each movie. Find the top 3 producers who have produced movies with the highest average RoI?

```python
#considering only those rows in crew column which have no null values
no_na_crew = imdb_df_new[~imdb_df_new['crew'].isna()]
```

```python
no_na_crew.shape
```

```python
# Functions
def create_producer_list(row):
    crew_list = row['crew']
    return [elem['name'] for elem in crew_list if elem['job'] == 'Producer']

def create_director(row):
    crew_list = row['crew']
    for elem in crew_list:
        if elem['job'] == 'Director':
            return elem['name']
    return None
```

```python
rows = []
for index, row in no_na_crew.iterrows():
    try:
        rows.append({
            'Movie Title': row['original_title'],
            'Producers': create_producer_list(row),
            'Directors': create_director(row),
            'ROI': row['roi']
        })
    except Exception as e:
        continue
```

```python
Table = pd.DataFrame(rows, columns=['Movie Title','Producers','Directors','ROI'])
```
